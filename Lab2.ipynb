```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from google.colab import drive

drive.mount('/content/drive')
```

```python
data = pd.read_csv('/content/drive/MyDrive/FMML/Datafiniti_Womens_Shoes.csv')
data = data[["brand","prices.color", "prices.amountMax", "prices.amountMin","prices.size" ]]
data.head()
```

```python
##Brand wise sales, can view the stats of sales of any particular Brand

brands = data["brand"].unique()
brand_sales = [data["brand"].tolist().count(brand) for brand in brands ]
plt.bar(brands,brand_sales)
plt.title("Brand wise sales")
plt.show()

##Color wise sales, can view the stats of sales of any particular color
##i don't know why it's not showing 

colors = data["prices.color"].unique()
color_sales = [data["prices.color"].tolist().count(color) for color in colors ]
plt.bar(colors, color_sales)
plt.title("color wise sales")
plt.show()

##size wise sales, can view the stats of sales of any particular size

sizes = data["prices.size"].unique()
size_sales = [data["prices.size"].tolist().count(size) for size in sizes ]
plt.bar(sizes, size_sales)
plt.title("size wise sales")
plt.show()


##through this plot, we can analyze which brand produced what colors the most

sns.scatterplot(data = data, x = "brand", y = "prices.color")

plt.show()

```
